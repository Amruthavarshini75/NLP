```python
from transformers import pipeline

# Initialize the summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Sample news article
article = """
The global economy is experiencing an unprecedented downturn as a result of the COVID-19 pandemic.
Governments around the world are implementing stimulus packages and relief measures to curb the economic impact,
but experts say recovery could take years. Meanwhile, unemployment rates are soaring, and many businesses are
struggling to stay afloat in these turbulent times.
"""

# Generate the summary
summary = summarizer(article, max_length=50, min_length=25, do_sample=False)

# Print the summary
print("Summary:", summary[0]['summary_text'])
```

```python
from summarizer import Summarizer

# Initialize the BERT summarizer
model = Summarizer()

# Sample news article
article = """
The global economy is experiencing an unprecedented downturn as a result of the COVID-19 pandemic.
Governments around the world are implementing stimulus packages and relief measures to curb the economic impact,
but experts say recovery could take years. Meanwhile, unemployment rates are soaring, and many businesses are
struggling to stay afloat in these turbulent times.
"""

# Generate the extractive summary
summary = model(article, ratio=0.2)  # Extract 20% of the most important sentences

# Print the summary
print("Summary:", summary)
```
